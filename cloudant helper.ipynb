```python
cloudant_url = "https://"+username + ":" + password + "@" + username + ".cloudant.com"
user_ctx = {
    "name": username,
    "roles": [
      "_admin",
      "_reader",
      "_writer"
    ]
  }
```

```python
from cloudant.client import Cloudant
from cloudant.replicator import Replicator
```

```python
client = Cloudant(username, password, url = cloudant_url, connect=True)

session = client.session()
```

```python
client = Cloudant(username, password, url = cloudant_url, connect=True)

session = client.session()
    
retaildemo_dept = client.create_database('retaildemo-dept')
retaildemo_desk = client.create_database('retaildemo-desk')
retaildemo_slot = client.create_database('retaildemo-slot')
retaildemo_visitor = client.create_database('retaildemo-visitor')
retaildemo_store = client.create_database('retaildemo-store')

rep_ins = Replicator(client)

dept = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-dept', target=cloudant_url + "/retaildemo-dept", repl_id = 'retaildemo-dept', user_ctx = user_ctx)

desk = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-desk', target=cloudant_url + "/retaildemo-desk", repl_id = 'retaildemo-desk', user_ctx = user_ctx)

slot = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-slot', target=cloudant_url + "/retaildemo-slot", repl_id ='retaildemo-slot', user_ctx = user_ctx)

visitor = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-visitor', target=cloudant_url + "/retaildemo-visitor", repl_id ='retaildemo-visitor',user_ctx = user_ctx)

store = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-store', target=cloudant_url + "/retaildemo-store", repl_id ='retaildemo-store', user_ctx = user_ctx)

print "Done!"
```

Now go back to the Replicatoin tab in your Cloudant instance, you will see a queue of replication tasks, wait until all of them are completed. You may need to click Refresh to see the updates.
