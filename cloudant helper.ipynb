```python
from cloudant.client import Cloudant
from cloudant.replicator import Replicator
```

```python
client = Cloudant(username, password, url = "https://"+username +".cloudant.com", connect=True)

session = client.session()
    
retaildemo_dept = client.create_database('retaildemo-dept')
retaildemo_desk = client.create_database('retaildemo-desk')
retaildemo_slot = client.create_database('retaildemo-slot')
retaildemo_visitor = client.create_database('retaildemo-visitor')
retaildemo_store = client.create_database('retaildemo-store')

rep_ins = Replicator(client)

dept = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-dept', target_db=client['retaildemo-dept'], repl_id = 'retaildemo-dept')

desk = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-desk', target_db=client['retaildemo-desk'], repl_id = 'retaildemo-desk')

slot = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-slot', target_db=client['retaildemo-slot'], repl_id ='retaildemo-slot')

visitor = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-visitor', target_db=client['retaildemo-visitor'], repl_id ='retaildemo-visitor')

store = Replicator.create_replication(rep_ins, source='https://redbookid.cloudant.com/retaildemo-store', target_db=client['retaildemo-store'], repl_id ='retaildemo-store')

print "Done!"
```

Now go back to the Replicatoin tab in your Cloudant instance, you will see a queue of replication tasks, wait until all of them are completed. You may need to click Refresh to see the updates.
